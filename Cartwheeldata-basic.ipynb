```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd 
import numpy as np 
```

```python
url = 'https://raw.githubusercontent.com/tamalmuk21/Stat_with_python/master/stat/Cartwheeldata.csv'
df = pd.read_csv(url)
df.head()
```

```python
sns.scatterplot(df)
```

```python
sns.scatterplot(x='Wingspan', y='CWDistance', data=df);
```

```python
#color according to output 
sns.scatterplot(data = df , x='Wingspan', y='CWDistance' , hue = 'Gender')
```
