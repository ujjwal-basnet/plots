```python
import seaborn as sns
```

<h2>ECDF stands for Empirical Commutative Distribution</h2>

-) In an ECDF plot, the x-axis corresponds to the range of values for the variable whereas the y-axis corresponds to the proportion of data points that are less than or equal to the corresponding value of the x-axis 


-) We can make the ECDF plot directly by using ecdfplot() function, or  using displot() function with the new Seaborn version

```python
from matplotlib import pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set_style(style = 'darkgrid' )
#creating ecdfplot 
sns.ecdfplot(data = tips , x = 'tip' )
```

```python
#show that above 85 percentage the tip is less then 5 
sns.ecdfplot(data = tips , x = 'tip' )
plt.axvline(x = 5  , color = "black"  )

```

```python
sns.ecdfplot(data = tips , x = 'tip' )
plt.axhline(y = 0.8 , xmin = 0 , xmax = 0.4 , color = 'black')
#note that xmax , xmin values should be between 0 to 1 
```

```python
#yaxis 
sns.ecdfplot (data = tips  , y = 'tip') 
#almost  80 percent proportion  gives less than 4 dollor tips 
```

```python
#hue 
sns.ecdfplot(x='tip', data=tips, hue='time');
#conclusion : Tips ammount is less in lunch compared to dinner 
# max tips in lunch is between 6 - 8 where as max tip on dinner is 10 dollor 
```

```python
sns.ecdfplot(x = 'tip' , data = tips , hue ='day' , stat = 'proportion #by default stat is proprotion we can convert into coount '  )
```

```python
#count 
sns.ecdfplot(data = tips , x = 'tip' , hue = 'day' , stat = 'count')
```

```python
#count 
sns.ecdfplot(data = tips , x = "tip" , stat = 'count')
```

```python
#Using  weights to weigh the proportion. 
```

```python
sns.ecdfplot(x='tip', data=tips, weights='tip')
plt.axhline(0.5, c='black');
```
